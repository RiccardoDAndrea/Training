# Leopar 2 A6

```python
# Step 1
from skimage import io, color

# Bild einlesen
img_Leo_2A6 = io.imread("Leopard_2A6.jpg")
io.imshow(img_Leo_2A6)

# Zeige das Bild an
#io.show()
```

```python
import matplotlib.pyplot as plt

# Nehmen wir an, img_Leo_2A6 ist Ihr Bild

# Extrahieren der einzelnen Farbkanäle
red_channel = img_Leo_2A6[:, :, 0]
green_channel = img_Leo_2A6[:, :, 1]
blue_channel = img_Leo_2A6[:, :, 2]

# Erstellen von drei separaten Plots für jeden Farbkanal
plt.figure(figsize=(12, 4))

# Plot für den roten Farbkanal
plt.subplot(1, 3, 1)
plt.hist(red_channel.ravel(), bins=256, color='red', alpha=0.5, label='Red Channel')
plt.title('Red Channel Histogram')

# Plot für den grünen Farbkanal
plt.subplot(1, 3, 2)
plt.hist(green_channel.ravel(), bins=256, color='green', alpha=0.5, label='Green Channel')
plt.title('Green Channel Histogram')

# Plot für den blauen Farbkanal
plt.subplot(1, 3, 3)
plt.hist(blue_channel.ravel(), bins=256, color='blue', alpha=0.5, label='Blue Channel')
plt.title('Blue Channel Histogram')

# Layout anpassen, um Überlagerungen zu vermeiden
plt.tight_layout()

# Plots anzeigen
plt.show()
```

```python
# Step 2
# Konvertiere das Bild zu Graustufen
img_Leo_2A6_gray = color.rgb2gray(img_Leo_2A6)

# Zeige das Bild in Graustufen an
io.imshow(img_Leo_2A6_gray, cmap='gray')

# Zeige das Bild an
io.show()
```

```python
# Step 3 setting a threshold

# Import the otsu threshold function
from skimage.filters import threshold_otsu

# Obtain the optimal otsu global thresh value
global_thresh = threshold_otsu(img_Leo_2A6_gray)

# Obtain the binary image by applying global thresholding
binary_global = img_Leo_2A6_gray > global_thresh

```

```python
# 4. Vis the picture without the Noise
# Show the binary image obtained
import matplotlib.pyplot as plt

# Deine vorherigen Import-Anweisungen und Code-Snippets

# Zeige das binäre Bild an
plt.imshow(binary_global, cmap='gray')
plt.title('Global thresholding')
plt.axis('off') 
plt.show()
```

```python
from skimage.filters import try_all_threshold
fig, ax = try_all_threshold(img_Leo_2A6_gray, verbose=False)
plt.figure(figsize = [30,20])
plt.show()
```

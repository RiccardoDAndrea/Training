```python
# Obejection Detection 
## 1. Import the Picture (jpg)
## 2. Create a Histogramm to see the channel (Colours)
## 2. Setting the Picture from RGB to grey
## 3. Setting the threshold
## 4. Vis the picture without the Noise
```

```python
# Step 1
from skimage import io, color

# Bild einlesen
img_Leo_2A6 = io.imread("Leopard_2A6.jpg")
io.imshow(img_Leo_2A6)

# Zeige das Bild an
#io.show()
```

```python
red_channel = img_Leo_2A6[:, :, 0]

# Plot the red histogram with bins in a range of 256
plt.hist(red_channel.ravel(), bins=256)

# Set title and show
plt.title('Red Histogram')
plt.show()
```

```python
# Step 2
# Konvertiere das Bild zu Graustufen
img_Leo_2A6_gray = color.rgb2gray(img_Leo_2A6)

# Zeige das Bild in Graustufen an
io.imshow(img_Leo_2A6_gray, cmap='gray')

# Zeige das Bild an
io.show()
```

```python
# Step 3 setting a threshold

# Import the otsu threshold function
from skimage.filters import threshold_otsu

# Obtain the optimal otsu global thresh value
global_thresh = threshold_otsu(img_Leo_2A6_gray)

# Obtain the binary image by applying global thresholding
binary_global = img_Leo_2A6_gray > global_thresh

```

```python
# 4. Vis the picture without the Noise
# Show the binary image obtained
import matplotlib.pyplot as plt

# Deine vorherigen Import-Anweisungen und Code-Snippets

# Zeige das binre Bild an
plt.imshow(binary_global, cmap='gray')
plt.title('Global thresholding')
plt.axis('off')  # Optional: Achsen ausschalten
plt.show()
```

```python
import matplotlib.pyplot as plt
from skimage import io
from skimage.filters import roberts, sobel, scharr, prewitt
from skimage.transform import rescale
```

```python
img = io.imread("/Users/riccardo/Desktop/Hubble_ultra_deep_field.jpg",as_gray=True)
img_rescaled = rescale(img, 1/4, anti_aliasing=False)
print(img_rescaled.shape, img.shape)

# plt.imshow(img,
#            cmap=plt.cm.gray, 
#            interpolation='nearest')

edge_roberts = roberts(img_rescaled)
edge_scharr = scharr(img_rescaled)
edge_sobel = sobel(img_rescaled)
edge_prewitt = prewitt(img_rescaled)
plt.imshow(edge_prewitt)


fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True,
                         figsize=(8, 8))
ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Original image')

ax[1].imshow(edge_roberts, cmap=plt.cm.gray)
ax[1].set_title('Roberts Edge Detection')

ax[2].imshow(edge_sobel, cmap=plt.cm.gray)
ax[2].set_title('Sobel')

ax[3].imshow(edge_scharr, cmap=plt.cm.gray)
ax[3].set_title('Scharr')

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
```
